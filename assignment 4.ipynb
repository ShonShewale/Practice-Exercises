##While Loop

```python
## example
count=0

while count<5:
  print(count)
  count=count+1
```

```python
## print 1 to 10 using while loop
count=0

while count<=10:
  print(count)
  count=count+1
```

```python
##Use a while loop to print: 10 9 8 ... 1
count=10

while True:
  print(count)
  count=count-1
  if count==0:
    break
```

```python
##Use a while loop to calculate the sum of numbers from 1 to 10.
count=0
sum=0

while count<=10:
  sum=sum+count

  count=count+1

print(sum)
```

```python
##Use a for loop to print all even numbers between 2 and 20.
count=0

while count<=20:
  if count%2==0:
    print(count)
  count=count+1
```

```python
##Take input n and print the multiplication table of n using a for loop.
a=int(input())
b=0
while b<=10:
  print(f"{a} x {b} = {a * b}")
  b=b+1
```

```python
##Use a for loop to print: 1 4 9 16 25 ... 100
count=0

while count<=100:
  print(f"square of {count}={count**2}")
  count=count+1
```
