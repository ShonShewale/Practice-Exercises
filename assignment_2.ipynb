```python
# 4. Write a program that takes a user’s age and checks if they are a child (<13), teenager (13–19), adult (20–59), or senior citizen (60+).
age=int(input())
if age<13:
  print("child")
elif age>13 and age<19:
  print("teenager")
elif age>20 and age<59:
  print("adult")
else:
  print("citizen")
```

```python
# 5. Input three numbers and use logical operators (and, or) in if/elif/else to find the largest number.
a=int(input())
b=int(input())
c=int(input())
if a>=b and a>=c:
  print("A is greater")
elif b>=a and b>=c:
  print("B is greater")
else:
  print("C is greater")
```

```python
# 8. Write a program to check if a given year is a leap year using leap year rules.
year=int(input())
if year % 400== 0:
  print("it is leap year")
elif year % 4==0 and year % 100!=0:
  print("it is leap year")
else:
  print("it is not leap year")
```
